```python
import joblib
import numpy as np

# =========================
# Load model + columns
# =========================
model = joblib.load("model.pkl")
columns = joblib.load("columns.pkl")

# =========================
# User input
# =========================
area = 800
bedrooms = 2
bathrooms = 1
location = "rural"

# =========================
# Base input
# =========================
input_data = {
    "area": area,
    "bedrooms": bedrooms,
    "bathrooms": bathrooms
}

# =========================
# Add missing columns (0)
# =========================
for col in columns:
    if col not in input_data:
        input_data[col] = 0

# =========================
# Encode location
# =========================
loc_col = f"location_{location}"

if loc_col in input_data:
    input_data[loc_col] = 1

# =========================
# Convert to model format
# =========================
final_input = np.array([list(input_data.values())])

# =========================
# Prediction
# =========================
prediction = model.predict(final_input)[0]

print("🏠 Predicted House Price:", prediction)
```
